```python
import pandas as pd
import numpy as np
from datetime import timedelta
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
user_sleep_2020 = pd.read_csv("C:\\Users\\ljy25\\etri\\data\\user_sleep_2020.csv")
user_sleep_2020
```

```python
user_sleep_2020 = user_sleep_2020.drop(['timezone'],axis=1)
user_sleep_2020
```

```python
user_sleep_2020['startDt'] = pd.to_datetime(user_sleep_2020['startDt'],unit='s') + timedelta(hours=9)
user_sleep_2020['endDt'] = pd.to_datetime(user_sleep_2020['endDt'],unit='s') + timedelta(hours=9)
user_sleep_2020['lastUpdate'] = pd.to_datetime(user_sleep_2020['lastUpdate'],unit='s') + timedelta(hours=9)
user_sleep_2020
```

```python
user_sleep_2020 = user_sleep_2020.drop('lastUpdate', axis=1)
user_sleep_2020
```

```python
user_sleep_2020['userId'].unique()
```

```python
user_sleep_2020['total_sleep'] = user_sleep_2020['lightsleepduration'] + user_sleep_2020['deepsleepduration'] + user_sleep_2020['remsleepduration']
user_sleep_2020
```

```python
plt.scatter(user_sleep_2020['total_sleep'] + user_sleep_2020['durationtosleep'] , user_sleep_2020['sleep_score'])
```

```python
plt.scatter(user_sleep_2020['total_sleep'] , user_sleep_2020['sleep_score'])
```

```python
user_sleep_2020.columns
```

```python
sleep_heatmap = user_sleep_2020.drop(columns=['userId', 'date', 'startDt', 'endDt'])
sleep_heatmap
```

```python
correlation_matrix = sleep_heatmap.corr()
correlation_matrix
```

```python
# 히트맵 그리기
plt.figure(figsize=(14, 12))
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm', cbar=True)
plt.title('Correlation Heatmap')
plt.show()
```
