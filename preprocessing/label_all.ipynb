```python
path = 'C:\\Users\\ljy25\\etri\\data\\user\\user01-06\\user01-06'
users = os.listdir('C:\\Users\\ljy25\\etri\\data\\user\\user01-06\\user01-06')
print("user_list : ", users)
```

```python
for user in users:
    for time in os.listdir(os.path.join(path, user)):
        file_pattern = os.path.join(path, user, time, "*_label.csv")
        print(file_pattern)
```

```python
# 병합할 데이터프레임 리스트
df_list = []

for user in users:
    user_path = os.path.join(path, user)
    if os.path.exists(user_path):
        for time in os.listdir(user_path):
            time_path = os.path.join(user_path, time)
            if os.path.exists(time_path):
                # _label.csv로 끝나는 모든 파일을 찾습니다.
                file_pattern = os.path.join(time_path, "*_label.csv")
                csv_files = glob.glob(file_pattern)
                # 모든 CSV 파일을 읽어들여 데이터프레임 리스트에 추가합니다.
                for file in csv_files:
                    df_list.append(pd.read_csv(file))

# 모든 데이터프레임을 하나로 병합합니다.
combined_df = pd.concat(df_list, ignore_index=True)

# 결과를 확인합니다.
combined_df
```
